```python

# Create a Bokeh figure
plot = figure(title="Hello Kitty Timeline",
              x_axis_label="Events",
              y_axis_label="Year",
              x_range=(-2, 2),
              y_range=(1970, 2025),
              width=200,
              height=800,
              tools="")

# Add circle glyphs for each data point (swap x and y coordinates)
circles = plot.circle(x=0, y="year", size=10, source=source, color=Category10[10][0], alpha=0.8)

# Add hover tool
hover = HoverTool(
    tooltips="""
    <div>
        <div>
            <img src="@image" alt="Hello Kitty Image" width="200" height="auto" border="2" style="float: left; margin: 0px 15px 15px 0px;"/>
        </div>
        <div>
            <span style="font-size: 17px; font-weight: bold;">@year</span>
            <br>
            <span style="font-size: 15px;">@event</span>
        </div>
    </div>
    """,
    renderers=[circles],
    mode="mouse"
)
plot.add_tools(hover)

# Remove gridlines and minor ticks
plot.xgrid.grid_line_color = None
plot.ygrid.grid_line_color = None
plot.xaxis.minor_tick_line_color = None
plot.yaxis.minor_tick_line_color = None

# Set the output file
# output_file("hello_kitty_timeline_vertical.html")

# # Show the result
# show(plot)


# from bokeh.io import save
# from bokeh.resources import CDN

# # Set the output file
# output_file("hello_kitty_timeline_vertical_centered.html")

# Add CSS to center the graph
# centered_html = f"""
# <!DOCTYPE html>
# <html lang="en">
#     <head>
#         <meta charset="utf-8">
#         <title>Hello Kitty Vertical Timeline</title>
#         {CDN.render()}
#         <style>
#             .center {{
#                 display: flex;
#                 justify-content: center;
#                 align-items: center;
#                 height: 100vh;
#             }}
#         </style>
#     </head>
#     <body>
#         <div class="center">
#             {plot.outer_html()}
#         </div>
#     </body>
# </html>
# """

# # Save the result
# with open("hello_kitty_timeline_vertical_centered.html", mode="w", encoding="utf-8") as f:
#     f.write(centered_html)

# # Show the result
# show(plot)

```

```python

# Set the output file
output_file("hello_kitty_timeline_vertical_centered.html")

# Get the script and div tags for embedding the plot
script, div = components(plot)

# Add CSS to center the graph
centered_html = f"""
<!DOCTYPE html>
<html lang="en">
    <head>
        <meta charset="utf-8">
        <title>Hello Kitty Vertical Timeline</title>
        {CDN.render()}
        <style>
            .center {{
                display: flex;
                justify-content: center;
                align-items: center;
                height: 100vh;
            }}
        </style>
    </head>
    <body>
        <div class="center">
            {div}
        </div>
        {script}
    </body>
</html>
"""

# Save the result
with open("hello_kitty_timeline_vertical_centered.html", mode="w", encoding="utf-8") as f:
    f.write(centered_html)

# Show the result
show(plot)
```
